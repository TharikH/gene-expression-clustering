```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
tasic_pc = pd.read_csv("zeisel_2016_pc.csv")
```

```python
X = np.array(tasic_pc.iloc[:, 1:])
```

```python
X[np.isnan(X)] = 0
```

```python
x = np.array(tasic_pc["PC1"])
y = np.array(tasic_pc["PC2"])
n_clusters = 11
k_means = KMeans(n_clusters=n_clusters, random_state=0, init='k-means++').fit(X)
```

```python
plt.scatter(x, y, c=k_means.labels_)
```

```python
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2).fit_transform(X)
```

```python
X_tsne.shape
x = X_tsne[:, 0]
y = X_tsne[:, 1]
plt.scatter(x, y, c = k_means.labels_)
```
